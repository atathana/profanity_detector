```python
from profanity_detector.get_data import get_all_movie_data,get_movie_locations_df
from geopy.geocoders import Nominatim
from imdb import Cinemagoer
```

```python
movie_data = get_all_movie_data ("Captain America")
```

```python
movie_location = get_movie_locations_df(movie_data)
```

```python
def split_location(location):
    return location.split(":")[0]
```

```python
geolocator = Nominatim(user_agent="profanity_detector")
```

```python
movie_location[0] = movie_location[0].apply(split_location)
```

```python
movie_location[0]
```

```python
def get_latitude(location):
    film_location = geolocator.geocode(location)
    if film_location is None:
        new_location = location.split(",")[-3:]
        film_location=geolocator.geocode(new_location)
    results = (film_location.latitude, film_location.longitude)
    return results[0]
movie_location["latitude"] = movie_location[0].apply(get_latitude)
```

```python
def get_longitude(location):
    film_location = geolocator.geocode(location)
    if film_location is None:
        new_location = location.split(",")[-3:]
        film_location=geolocator.geocode(new_location)
    results = (film_location.latitude, film_location.longitude)
    return results[1]
movie_location["longitude"] = movie_location[0].apply(get_longitude)
```

```python
movie_location
```

```python
def get_coordinates(location):
    film_location = geolocator.geocode(location)
    if film_location is None:
        new_location = location.split(",")[-3:]
        film_location=geolocator.geocode(new_location)
    results = (film_location.latitude, film_location.longitude)
    return results
movie_location["latitude"], movie_location["longitude"] = zip(*map(get_coordinates, movie_location[0]))
```
